# Crisp-DM Model

* Cross Industry Standard Process for Data Mining
* Iteration is the rule rather than the exception
  * Going through the process without having solved the problem is not a failure!
  * Often: Entire process is an exploration of data

# Business Understanding
* It is vital to understand the problem to be solved
* Is an iterative process of recasting the problem and designing a solution -> Cycles within a cycle
* Formulation of the business problem as one or more data science problems
  * Decomposing business problem into (Data Science) Sub Tasks
  * What could solve the problem? Classification, Regression, Propability estimiation?
* Skills needed:
 * Talk the same language as the business
 * Decompose problem into tasks for which tools (e.g. Python) are available

# Use Case thinking
* What exactly do we want to do?
* How exactly do we want to do it?
* What parts of this use scenario constitute which Machine Learning algorithm?
* What does the Business want?
 * Presentation?
 * APP?
 * Development of Machine Learning system? 
* What are the expectations?
* How can we measure our solution (Expected Value)?

# Example (Draft)
* Churn Problem at Telecommunication Company
* Subtask to solve the problem:
 * Estimate from historical data the probability of a customer terminating her contract shortly after it has expired
 * Get the data in specific format
 * Develop "model" to learn from the past

# Data Understanding
* How do we solve this goal? DATA!
* Important to know the strenghts and limitations of data because rarely is there an exact match with the problem
* Customer Database, Transaction Database, marketing response database all contain different information, different intersecting populations and may have varying degrees of reliability
* Estimation of "costs" of data: What data do we have? How much do we have to pay for new data? How much does it cost to clean the data?
* Dig beneath the surface to uncover the structure of a business problem
* Skills needed:
 * Data & Databases (streaming data, relational data, csv, sql, Python ...)
 * Data Visualization (Plots)
 * Statistics

# Data Preparation
* Analytic technologies for Data Understanding impose certain requirements on the data they use
* Often different form from how the data is provided naturally 
* Data needs to be transformed or converted
* Skills needed:
 * Regular expressions
 * Data & Databases
 * Data manipulation

# Examples
* Converting data to tabular format
* removing or inferring missing values
* Converting data to different types (categorial data, numeric values)

# Modeling
* Output of modeling is some sort of model or pattern capturing regularities in the data
* Primary place of Machine Learning Models
* Skills needed:
 * Machine Learning
 * Programming
 * Math

# Evaluation
* Assess the (Machine Learning) results
* Gain confidence that they are valid and reliable before moving on
* If you looks too hard on a dataset you will find patterns!
* Also, you might have luck with the (test-) dataset
* Does the model satisfy the original business goals?
* Includes quantitative and qualitative assessments
* Skills needed:
 * Statistics
 * Business Understanding
 * Communication
